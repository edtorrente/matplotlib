## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combine_data_df = pd.merge(mouse_metadata, study_results, on='Mouse ID', how='outer' )

# Display the data table for preview
combine_data_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
remove_dup_df = index_df.drop(['g989'])
#determine if the duplicates of '989' are removed.
remove_dup_df.shape
```

```python
# Checking the number of mice in the clean DataFrame.
clean_df = remove_dup_df.reset_index()
clean_df['Mouse ID'].value_counts()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:

regimen_group = clean_df.groupby(['Drug Regimen'])
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
tumor_vol_mean = round (regimen_group['Tumor Volume (mm3)'].mean(), 2)
tumor_vol_med = round (regimen_group['Tumor Volume (mm3)'].median(), 2)
tumor_vol_var = round (regimen_group['Tumor Volume (mm3)'].var(), 2)
tumor_vol_std = round (regimen_group['Tumor Volume (mm3)'].std(), 2)
tumor_vol_sem = round (regimen_group['Tumor Volume (mm3)'].sem(), 2)

# Assemble the resulting series into a single summary dataframe.

tumor_vol_df = pd.DataFrame({'Mean': tumor_vol_mean, 'Median': tumor_vol_med, 'Variance': tumor_vol_var,
                            'Standard Deviation': tumor_vol_std, 'Std. Err of Mean': tumor_vol_sem})

tumor_vol_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
tumor_vol_2 = clean_df.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].agg(['mean', 'median', 'var','std', 
                                                                            'sem'])
tumor_vol_2
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of timepoints for all mice tested for each drug regimen using 
# Pandas.

total_timepoints = clean_df.groupby(['Drug Regimen'])['Timepoint'].count().sort_values(ascending = False)
bar_graph_pandas = total_timepoints.plot(kind='bar', color='orange', figsize = (10,5))
bar_graph_pandas.set_title('Drug Count')
bar_graph_pandas.set_xlabel('Drug')
bar_graph_pandas.set_ylabel('Count')
plt.show()
```

```python
# Generate a bar plot showing the total number of timepoints for all mice tested for each drug regimen using 
# pyplot.

total_timepoints_sort = total_timepoints.sort_values(ascending = False) #need to sort in descending values
#use index to sort drug name and values to find values for each drug
plt.bar(total_timepoints_sort.index, total_timepoints_sort.values, color='orange', width=0.5)
plt.xlabel('Drug Regimen')
plt.ylabel('Counts', size=10)
plt.title('Drug Count')
plt.xticks(rotation='vertical')

plt.show()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using Pandas

sex_dist = clean_df.groupby(['Sex'])['Sex'].count()
sex_dist_pie = sex_dist.plot(kind ='pie', y=sex_dist, colors = {'orange', 'blue'}, autopct="%2.0f%%", 
                             shadow=True, startangle=180 )
sex_dist_pie.set_title('Gender')
sex_dist_pie.set_ylabel

plt.show()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pyplot

plt.pie(sex_dist, labels=('female', 'male'), startangle=180, autopct='%2.0f%%')
plt.title('Gender')
plt.ylabel('Sex')
plt.tight_layout()
plt.show()
```

## Line and Scatter Plots

```python
#generate a list to chose which MOUSE ID to plot

capomulin_drug = clean_df.loc[clean_df['Drug Regimen'] == 'Capomulin']
mouse_treated = capomulin_drug[['Mouse ID']].drop_duplicates(subset = 'Mouse ID', keep ='last')
mouse_treated
```

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
# choose the mouse of interest.
mouse_r157 = clean_df.loc[clean_df['Mouse ID'] == 'r157']
x_axis = mouse_r157['Timepoint']
y_axis = mouse_r157['Tumor Volume (mm3)']
plt.plot(x_axis, y_axis, color='red', marker='o', markerfacecolor = 'blue', linewidth =3)
plt.title('Mouse r157 Capmulin Treatment')
plt.xlabel ('Timepoint (days)')
plt.ylabel ('Tumor Vol (mm3)')
plt.show()
```

```python
# create dataframe and use loc function for the Capomulin drug
capomulin_df = clean_df.loc[clean_df['Drug Regimen'] == 'Capomulin']
#use groupby and mean function to determine average
capomulin_avg = capomulin_df.groupby(['Mouse ID']).mean()
capomulin_avg
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen

plt.scatter(capomulin_avg['Weight (g)'], capomulin_avg['Tumor Volume (mm3)'], facecolor='red')
plt.xlabel('Weight(g)')
plt.ylabel('Tumor Volume (mm3)')
plt.title('Mouse Weight vs. Tumor Volume')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient for mouse weight and average tumor volume for the Capomulin regimen

# use the pearson formula and the average calculated above to determin p-value.
# if p-value is close to 1, there is a high degree of correlation between mouse weight and average tumor volume
# if p-value is close to 0, there is no correlation
# if p-value is close to -1 suggests there is negative correlation

correlation_coeff = round(st.pearsonr(capomulin_avg['Weight (g)'], capomulin_avg['Tumor Volume (mm3)'])[0], 4)
correlation_coeff
```

```python
#we can find the a linear model using st.linregression
linear_model = st.linregress(capomulin_avg['Weight (g)'], capomulin_avg['Tumor Volume (mm3)'])
linear_model
```

```python
#linear regression model for mouse weight and average tumor volume for the Capomulin regimen
slope=0.9544396890241049
intercept=21.552160532685008

regression_model = slope*(capomulin_avg['Weight (g)']) + intercept
plt.scatter(capomulin_avg['Weight (g)'], capomulin_avg['Tumor Volume (mm3)'])
plt.plot(capomulin_avg['Weight (g)'], regression_model, 'r', label='{:.4f}x+{:.4f}'.format(slope, intercept))
plt.legend()
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()

```
